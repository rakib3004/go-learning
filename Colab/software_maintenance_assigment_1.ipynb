# Calculate SLOC for a Go Language Source Code

**Task 1:** You are given a source code file of a specific programming language. You have to calculate the SLOC (only source line of code without comment) of that file.


```python
def count_sloc(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    sloc_count = 0
    in_block_comment = False

    for line in lines:
        line = line.strip()

        if line.startswith('/*'):
            in_block_comment = True
        elif line.endswith('*/'):
            in_block_comment = False
        elif line.startswith('//'):
            continue
        elif not in_block_comment and line:
            sloc_count += 1

    return sloc_count

# Replace 'your_go_file.go' with the actual file path
file_path = '/content/sample_data/Task1/function.go'
sloc = count_sloc(file_path)
print("SLOC (Source Lines of Code) without comments:", sloc)
```

**Task 2:** You will be given two versions of a source code file. You have to identify the number of changes (addition, deletion, total changes) between these two files.


```python
import difflib
import pandas as pd

def compare_versions(file1_content, file2_content):
    diff = difflib.ndiff(file1_content, file2_content)
    additions = 0
    deletions = 0

    for line in diff:
        if line.startswith('+ '):
            additions += 1
        elif line.startswith('- '):
            deletions += 1

    total_changes = additions + deletions

    return additions, deletions, total_changes

# Replace 'version1_file_path' and 'version2_file_path' with the actual file paths
version1_file_path = '/content/sample_data/Task2/initial.go'
version2_file_path = '/content/sample_data/Task2/updated.go'


with open(version1_file_path, 'r') as file:
    version1_content = file.readlines()

with open(version2_file_path, 'r') as file:
    version2_content = file.readlines()

additions, deletions, total_changes = compare_versions(version1_content, version2_content)

# Creating a Pandas DataFrame to display the changes
data = {
    'Type': ['Additions', 'Deletions', 'Total Changes'],
    'Count': [additions, deletions, total_changes]
}

changes_df = pd.DataFrame(data)

# Display the changes in a table
changes_df
```

**Task 3:** You will be given two versions of source code repository (two folders which contain source code at two different versions). You have to identify which files have been changed and calculate the number of changes (addition, deletion, total changes).

```python
import difflib
import os
import pandas as pd

def compare_files(file1_content, file2_content):
    diff = difflib.ndiff(file1_content, file2_content)
    additions = 0
    deletions = 0

    for line in diff:
        if line.startswith('+ '):
            additions += 1
        elif line.startswith('- '):
            deletions += 1

    total_changes = additions + deletions

    return additions, deletions, total_changes

def process_file(file_path, version1_path, version2_path):
    with open(version1_path + file_path, 'r') as file1:
        version1_content = file1.readlines()

    with open(version2_path + file_path, 'r') as file2:
        version2_content = file2.readlines()

    additions, deletions, total_changes = compare_files(version1_content, version2_content)
    return additions, deletions, total_changes

def process_folder(version1_path, version2_path):
    file_changes = []

    for root, _, files in os.walk(version1_path):
        for file in files:
            file_path = os.path.join(root, file)
            additions, deletions, total_changes = process_file(file_path[len(version1_path):], version1_path, version2_path)
            file_changes.append({
                'File': file_path[len(version1_path):],
                'Additions': additions,
                'Deletions': deletions,
                'Total Changes': total_changes
            })

    changes_df = pd.DataFrame(file_changes)
    return changes_df

# Replace 'version1_path' and 'version2_path' with the actual paths of the two versions' folders
version1_path = '/content/sample_data/Task3/version1'
version2_path = '/content/sample_data/Task3/version2'

changes_df = process_folder(version1_path, version2_path)

# Display the changes in a table using Pandas DataFrame
changes_df
```
